```python
import pandas as pd
df = pd.read_csv('shopping_behavior_updated.csv')
df.head()
```

```python
df.describe(include='all')
```

```python
df_partial = pd.DataFrame({'Age': df['Age'], 'Gender': df['Gender'], 'Location': df['Location'], 'Item Purchased': df['Item Purchased']})
df_partial.head()
```

```python
Y=df_partial['Gender']
X=df_partial.drop('Gender', axis=1)
Y.head()
```

```python
locations = df_partial['Location'].unique()
codici = list(range(50))
print(locations)
```

```python
location_map = dict(zip(locations, codici))
print(location_map)
locs = df_partial['Location']
locs = locs.map(lambda x: location_map.get(x))
print(locs)
```

```python
items = df_partial['Item Purchased'].unique()
codici_items = list(range(25))
#print(items)
items_map = dict(zip(items, codici_items))
oggetti = df_partial['Item Purchased']
oggetti = oggetti.map(lambda x: items_map.get(x))
print(items_map)
```

```python
genders = Y.unique()
print(genders)
gender_map = {'Male': 0, 'Female': 1}
#genders = df_partial['Gender']
Y = Y.map(gender_map)
X['Item Purchased'] = oggetti
X['Location'] = locs
#Y['Gender'] = genders
print(Y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

X_train.head()
Y_train.head()
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

print(f"Accuracy:{accuracy_score(Y_test, y_pred):.2f}")
#print(f"Precision:{precision_score(Y_test, y_pred):.2f}")
#print(f"Recall:{recall_score(Y_test, y_pred):.2f}")
#print(f"F1:{f1_score(Y_test, y_pred):.2f}")
```
